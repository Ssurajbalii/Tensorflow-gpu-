```python
import turicreate as tc
import numpy as np
import matplotlib.pyplot as plt
import utils
```

```python

data = tc.SFrame('Hyderabad.csv')

data
```

```python
tc.plot(data['Area'],data['Price'] ,'Area','Price','First plot using Turicreate')
```

```python
data['Price'].explore()
```

```python
tc.show(data['Price'],data['Area'] ,'Price','Area','First plot using Turicreate')
```

```python
tc.plot(data['Price'],data['No. of Bedrooms'],'Price','No. of Bedrooms','Second Plot using Turicreate')
```

```python
two_feature_model = tc.linear_regression.create(data,target = 'Price' , features = ['Area','No. of Bedrooms'])
```

```python
two_feature_model.coefficients
```

```python
two_feature_model.evaluate(data)
```

```python
house = tc.SFrame({'Area': [4000,1500,1000], 'No. of Bedrooms':[5,3,2]})
house
```

```python
two_feature_model.predict(house)
```
